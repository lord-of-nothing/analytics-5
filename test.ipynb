```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

# Небольшие изменения в датасете

```python
df = pd.read_csv('../datasets/group2_small_final.csv')
```

```python
df['date'] = df['timestamp'].str[:10]
```

```python
df['category'].unique()
```

# Прогноз по выручке на следующий год

```python
p = Prophet()
df1 = df[['value', 'date']]
df1.columns = ['y', 'ds']
df1 = df1.groupby('ds').agg({'y': 'sum'}).reset_index()
p.fit(df1)
future = p.make_future_dataframe(periods=363, include_history=False)
forecast = p.predict(future)
forecast = forecast[['ds', 'yhat']]
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(forecast['ds'], forecast['yhat'], linewidth=1.5, color='red')
```

# Прогнозы на спрос пользователей к категориям

## Прогноз по категории Красота и Здоровье

```python
df2 = df[(df['action'] == 'cart') & (df['category'] == 'Красота и здоровье')].groupby('date').agg(
    {'value': 'count'}).reset_index()
df2.columns = ['ds', 'y']
p1 = Prophet()
p1.fit(df2)
future1 = p1.make_future_dataframe(periods=363, include_history=False)
forecast1 = p1.predict(future1)
forecast1 = forecast1[['ds', 'yhat']]
fig1, ax = plt.subplots(figsize=(15, 5))

ax.plot(forecast1['ds'], forecast1['yhat'], linewidth=1.5, color='green')
```
